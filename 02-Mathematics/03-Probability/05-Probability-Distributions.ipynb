# Probability Distributions

Probability distributions are a fundamental concept in probability theory and statistics, and they are used to model the behavior of random variables. There are two main types of probability distributions:

- **Discrete Distributions:** Discrete distributions refer to probability distributions that arise from countable and distinct outcomes. In other words, the random variable can only take on specific values, and there are no intermediate values between them. Examples of discrete distributions include the binomial distribution, Poisson distribution, and geometric distribution.

- **Continuous Distributions:** Continuous distributions, on the other hand, refer to probability distributions that arise from uncountable and continuous outcomes. In other words, the random variable can take on any value within a certain range, and there can be an infinite number of intermediate values between any two points. Examples of continuous distributions include the normal distribution, uniform distribution, and exponential distribution.

In this response, we will discuss some of the most commonly used probability distributions.

## Binomial Distribution

The binomial distribution is used to model the number of successes in a fixed number of independent trials. Each trial has only two possible outcomes, usually called success and failure. The binomial distribution is characterized by two parameters: the probability of success in each trial `(p)` and the number of trials `(n)`.

`Example:` A coin is tossed 10 times. What is the probability of getting exactly 4 heads?

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

n, p = 10, 0.5 # number of trials, probability of success
k = np.arange(0, n+1) # range of possible number of successes
binomial = binom.pmf(k, n, p) # probability mass function

plt.plot(k, binomial, 'o--')
plt.title('Binomial Distribution')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.show()
```

## Poisson Distribution

The Poisson distribution is used to model the number of rare events occurring in a fixed interval of time or space. It is characterized by a single parameter, usually denoted `Î»`, which represents the expected number of events in the interval.

`Example:` The average number of cars that pass through a particular intersection per hour is 6. What is the probability that exactly 8 cars will pass through the intersection in a given hour?

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

mu = 6 # mean of the Poisson distribution
k = np.arange(0, 20) # range of possible values
poisson_dist = poisson.pmf(k, mu) # probability mass function

plt.plot(k, poisson_dist, 'o--')
plt.title('Poisson Distribution')
plt.xlabel('Number of Events')
plt.ylabel('Probability')
plt.show()
```

## Uniform Distribution

The uniform distribution is a continuous probability distribution that has a constant probability density function over a specified interval. The probability density function is flat and equal over the entire interval.

`Example:` A spinner has an equal chance of landing on any number between 1 and 6. What is the probability of it landing on 3?

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform

a, b = 1, 6 # range of possible values
x = np.linspace(a, b, 100)
uniform_dist = uniform.pdf(x, a, b-a) # probability density function

plt.plot(x, uniform_dist)
plt.title('Uniform Distribution')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.show()
```

## Normal Distribution

The normal distribution is a continuous probability distribution that is widely used in statistics to model a variety of natural phenomena. It is characterized by two parameters: the mean `(Î¼)` and the standard deviation `(Ï)`.

`Example:` The average height of students in a class is 170 cm with a standard deviation of 5 cm. What is the probability that a randomly selected student from the class will have a height between 160 cm and 180 cm?

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

mu, sigma = 170, 5 # mean and standard deviation
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
normal_dist = norm.pdf(x, mu, sigma) # probability density function

plt.plot(x, normal_dist)
plt.title('Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.show()
```

## Exponential Distribution

The exponential distribution is used to model the time between two consecutive events in a Poisson process. It is characterized by a single parameter, usually denoted `Î»`, which represents the rate at which events occur.

`Example:` The time (in minutes) between arrivals of customers at a coffee shop follows an exponential distribution with a mean of 5 minutes. What is the probability that the time between arrivals of two customers will be less than 3 minutes?

To solve this problem, we can use the cumulative distribution function (CDF) of the exponential distribution, which is given by:

\begin{equation}
F(x) = 1 - e^{-Î»x}
\end{equation}

where `Î»` is the rate parameter (the inverse of the mean), and `x` is the value of the random variable.

In our example, `Î» = 1/5`, so the CDF is:

\begin{equation}
F(x) = 1 - e^{-Î»x}
\end{equation}

To find the probability that the time between arrivals of two customers will be less than 3 minutes, we need to evaluate the CDF at `x = 3`:

\begin{equation}
P(X < 3) = F(3) = 1 - e^{-3/5} â 0.7769
\end{equation}

This means that there is a 77.69% chance that the time between arrivals of two customers will be less than 3 minutes.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon

# Set the rate parameter
Î» = 1/5

# Create an array of x values
x = np.linspace(0, 20, 1000)

# Calculate the PDF and CDF at each x value
pdf = expon.pdf(x, scale=1/Î»)
cdf = expon.cdf(x, scale=1/Î»)

# Plot the PDF and CDF
fig, ax = plt.subplots(2, 1, figsize=(8, 8))
ax[0].plot(x, pdf, label='PDF')
ax[0].set_xlabel('x')
ax[0].set_ylabel('f(x)')
ax[0].legend()

ax[1].plot(x, cdf, label='CDF')
ax[1].set_xlabel('x')
ax[1].set_ylabel('F(x)')
ax[1].legend()

plt.show()
```

This will plot the PDF and CDF of the exponential distribution with a mean of `5` minutes. The PDF will have a decreasing exponential shape, while the CDF will start at `0` and approach `1` as `x` increases.

## Skewness and Kurtosis

Skewness and kurtosis are measures of the shape of a probability distribution.

Skewness measures the degree of asymmetry in the distribution. A distribution is symmetric if its right and left sides are mirror images of each other. If a distribution has a longer tail on the right side, it is said to be positively skewed, while if it has a longer tail on the left side, it is said to be negatively skewed.

Kurtosis measures the degree of peakedness in the distribution. A distribution with a higher peak and heavier tails has positive kurtosis, while a flatter distribution with lighter tails has negative kurtosis.

These measures can be useful in comparing different distributions and identifying deviations from the normal distribution, which has zero skewness and kurtosis of 3.

Both skewness and kurtosis can be calculated using moments of the probability distribution.
