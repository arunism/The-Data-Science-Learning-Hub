# Matrix Mappings

A matrix mapping, also known as a matrix transformation or linear transformation, is a mapping that is defined by a matrix. It is a specific type of linear mapping that transforms vectors by matrix multiplication.

Mathematically, given a matrix $A$ of size $m x n$, a matrix mapping $T: R^n -> R^m$ is defined as:

$T(x) = A * x$

where $x$ is a vector in $R^n$ and $T(x)$ is the transformed vector in $R^m$.

In this context, the matrix $A$ represents the linear transformation and determines how the vectors are transformed. Each column of the matrix $A$ represents the image of the corresponding basis vector under the transformation.

The matrix mapping has the following properties:

- **Preserves vector addition:** $T(u + v) = T(u) + T(v)$
- **Preserves scalar multiplication:** $T(ku) = kT(u)$

Matrix mappings can be used to perform various transformations, such as rotations, scaling, shearing, and reflections, by appropriately choosing the matrix $A$.

For example, consider a matrix mapping that rotates a 2D vector counterclockwise by an angle $θ$. The matrix representation for this transformation is:

$A = [cos(θ) -sin(θ)]$

$[sin(θ) cos(θ)]$

To apply this matrix mapping to a vector $x = [x1, x2]$, you simply multiply the vector $x$ by the matrix $A$:

$T(x) = A * x$

This matrix mapping will rotate the vector $x$ by an angle $θ$.

Matrix mappings have a wide range of applications in various fields, including computer graphics, computer vision, physics, and engineering. They provide a powerful tool for expressing and performing transformations on vectors and are fundamental in linear algebra.

## Linear mappings

A linear mapping, also known as a linear transformation, is a mapping between two vector spaces that preserves vector addition and scalar multiplication. It maps lines to lines and the origin to the origin. Mathematically, a mapping $T: V → W$ is linear if it satisfies the following properties:

- $T(u + v) = T(u) + T(v)$ for all vectors $u$ and $v$ in $V$ (Preserves vector addition).
- $T(ku) = kT(u)$ for all vectors $u$ in $V$ and scalar $k$ (Preserves scalar multiplication).

- **Identity mapping:** Maps each vector to itself.
- **Zero mapping:** Maps each vector to the zero vector.
- **Scalar multiplication:** Multiplies each vector by a scalar.
- **Addition:** Adds two vectors component-wise.
- **Projection:** Projects a vector onto another vector or subspace.
- **Rotation:** Rotates a vector or set of vectors in a given plane or space.
- **Shearing:** Distorts the shape of an object along a particular direction.
- **Reflection:** Reflects a vector or set of vectors across a line, plane, or hyperplane.
- **Scaling:** Scales a vector or set of vectors uniformly.
- **Translation:** Shifts a vector or set of vectors by a constant offset.

`Example:` Scaling Transformation

Consider a scaling transformation that scales a 2D vector by a factor of 2. It stretches or shrinks the vector while maintaining its direction.

```python
import numpy as np
import matplotlib.pyplot as plt

# Original vector
v = np.array([1, 1])

# Scaling factor
scale_factor = 2

# Linear mapping
T = scale_factor * v

# Plotting the original vector and the transformed vector
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='b', label='Original Vector')
plt.quiver(0, 0, T[0], T[1], angles='xy', scale_units='xy', scale=1, color='r', label='Transformed Vector')
plt.xlim(-4, 4)
plt.ylim(-4, 4)
plt.axhline(0, color='k', linestyle='--')
plt.axvline(0, color='k', linestyle='--')
plt.grid(True)
plt.legend()
plt.title('Scaling Transformation')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```
