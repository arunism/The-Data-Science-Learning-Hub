## Types of Vectors:

**1. Geometric Vectors:**

Geometric vectors are often represented as arrows in space. They have both magnitude (length) and direction. Geometric vectors can be two-dimensional (2D) or three-dimensional (3D) and are commonly denoted using variables with arrows on top, such as $\vec{ùêö}$, $\vec{b}$, or $\vec{c}$. These vectors can be added, subtracted, and multiplied by scalars.

`Example:` Let's consider two 2D geometric vectors, ùêö = [3, 4] and ùêõ = [1, 2]. We can represent them as arrow diagrams.

**2. Polynomial Vectors:**

Polynomials can also be represented as vectors. Each term in a polynomial corresponds to an element in the vector, with the exponent representing the index.

`Example:` The polynomial ùëù(ùë•) = 3ùë•¬≤ + 2ùë• - 5 can be represented as the vector ùëù = [3, 2, -5].

```python
import numpy as np
import matplotlib.pyplot as plt

# Geometric Vector Plot
a = np.array([3, 4])  # Vector a
b = np.array([1, 2])  # Vector c
c = a - b  # Vector b

origin = np.array([0, 0])

plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1) # Plotting vectors
plt.quiver(*origin, *a, color='r', angles='xy', scale_units='xy', scale=1) # Plotting vector a
plt.text(a[0], a[1], 'a', ha='right', va='bottom', color='r')
plt.quiver(*origin, *b, color='b', angles='xy', scale_units='xy', scale=1) # Plotting vector b
plt.text(b[0], b[1], 'b', ha='left', va='top', color='b')
plt.quiver(*b, *c, color='g', angles='xy', scale_units='xy', scale=1, alpha=0.2) # Plotting vector c
plt.text(c[0], c[1], 'c', ha='right', va='bottom', color='g')
plt.quiver(*origin, *c, color='b', angles='xy', scale_units='xy', scale=1) # Plotting transformed vector b
plt.xlim(-1, 4)
plt.ylim(-1, 5)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Geometric Vector')
plt.grid()

# Polynomial Function Plot
x = np.linspace(-5, 5, 100)
y1 = x**2 - 3*x + 2
y2 = -2*x**2 + 4*x - 1

# Plotting polynomial function y1
plt.subplot(1, 2, 2)
plt.plot(x, y1, label='y = x^2 - 3x + 2')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Polynomial Function 1')
plt.grid()

# Plotting polynomial function y2
plt.plot(x, y2, label='y = -2x^2 + 4x - 1')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Polynomial Vectors')
plt.legend()
plt.grid()

# Adjust subplot spacing
plt.tight_layout()

# Show the plot
plt.show()
```

## Zero Vector, Unit Vector, and Sparse Vector

- **Zero Vector:** The zero vector, denoted as $\vec{0}$ or ùüé, has all its elements equal to zero. It has no direction but is often used as a reference point.
- **Unit Vector:** A unit vector has a magnitude of 1. It represents a direction but doesn't have any specific length.
- **Sparse Vector:** A sparse vector is a vector that contains mostly zero elements, except for a few non-zero elements.

`Example:`
- Consider a 3D vector ùêØ = [0, 0, 0]. This vector is the zero vector, as all its elements are zero.
- A unit vector can be obtained by dividing any non-zero vector by its magnitude. For instance, if ùêÆ = [2, 2] is a 2D vector, its unit vector would be ùêÆÃÇ = [0.707, 0.707] (approximately).
- A sparse vector could be ùë† = [0, 0, 3, 0, 0, 0, 0, 2], where only two elements are non-zero.
