## Types of Vectors:

**1. Geometric Vectors:**

Geometric vectors are often represented as arrows in space. They have both magnitude (length) and direction. Geometric vectors can be two-dimensional (2D) or three-dimensional (3D) and are commonly denoted using variables with arrows on top, such as $\vec{ùêö}$, $\vec{b}$, or $\vec{c}$. These vectors can be added, subtracted, and multiplied by scalars.

`Example:` Let's consider two 2D geometric vectors, ùêö = [3, 4] and ùêõ = [1, 2]. We can represent them as arrow diagrams.

**2. Polynomial Vectors:**

Polynomials can also be represented as vectors. Each term in a polynomial corresponds to an element in the vector, with the exponent representing the index.

`Example:` The polynomial ùëù(ùë•) = 3ùë•¬≤ + 2ùë• - 5 can be represented as the vector ùëù = [3, 2, -5].

```python
import numpy as np
import matplotlib.pyplot as plt

# Geometric Vector Plot
a = np.array([3, 4])  # Vector a
b = np.array([1, 2])  # Vector c
c = a - b  # Vector b

origin = np.array([0, 0])

plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1) # Plotting vectors
plt.quiver(*origin, *a, color='r', angles='xy', scale_units='xy', scale=1) # Plotting vector a
plt.text(a[0], a[1], 'a', ha='right', va='bottom', color='r')
plt.quiver(*origin, *b, color='b', angles='xy', scale_units='xy', scale=1) # Plotting vector b
plt.text(b[0], b[1], 'b', ha='left', va='top', color='b')
plt.quiver(*b, *c, color='g', angles='xy', scale_units='xy', scale=1, alpha=0.2) # Plotting vector c
plt.text(c[0], c[1], 'c', ha='right', va='bottom', color='g')
plt.quiver(*origin, *c, color='b', angles='xy', scale_units='xy', scale=1) # Plotting transformed vector b
plt.xlim(-1, 4)
plt.ylim(-1, 5)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Geometric Vector')
plt.grid()

# Polynomial Function Plot
x = np.linspace(-5, 5, 100)
y1 = x**2 - 3*x + 2
y2 = -2*x**2 + 4*x - 1

# Plotting polynomial function y1
plt.subplot(1, 2, 2)
plt.plot(x, y1, label='y = x^2 - 3x + 2')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Polynomial Function 1')
plt.grid()

# Plotting polynomial function y2
plt.plot(x, y2, label='y = -2x^2 + 4x - 1')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Polynomial Vectors')
plt.legend()
plt.grid()

# Adjust subplot spacing
plt.tight_layout()

# Show the plot
plt.show()
```

## Zero Vector, Unit Vector, and Sparse Vector

- **Zero Vector:** The zero vector, denoted as $\vec{0}$ or ùüé, has all its elements equal to zero. It has no direction but is often used as a reference point.
- **Unit Vector:** A unit vector has a magnitude of 1. It represents a direction but doesn't have any specific length.
- **Sparse Vector:** A sparse vector is a vector that contains mostly zero elements, except for a few non-zero elements.

`Example:`
- Consider a 3D vector ùêØ = [0, 0, 0]. This vector is the zero vector, as all its elements are zero.
- A unit vector can be obtained by dividing any non-zero vector by its magnitude. For instance, if ùêÆ = [2, 2] is a 2D vector, its unit vector would be ùêÆÃÇ = [0.707, 0.707] (approximately).
- A sparse vector could be ùë† = [0, 0, 3, 0, 0, 0, 0, 2], where only two elements are non-zero.

## Vector Dimensions

The dimension of a vector refers to the number of components it has. The most common vector dimensions are one-dimensional, two-dimensional, and three-dimensional vectors.

1. **One-Dimensional Vectors:** A one-dimensional vector, also known as a scalar, represents a single value.
It has only one component and can be thought of as a point on a number line. `Examples`: Temperature, time, speed, etc.

2. **Two-Dimensional Vectors:** A two-dimensional vector lies in a plane and requires two components to represent its magnitude and direction. It is commonly represented as an arrow in a Cartesian coordinate system. `Examples`: Displacement in a plane, velocity in two dimensions, etc.

3. **Three-Dimensional Vectors:** A three-dimensional vector exists in 3D space and requires three components to represent its magnitude and direction. It can be represented as an arrow with its tail at the origin in a three-dimensional Cartesian coordinate system. `Examples`: Position in space, force, velocity in three dimensions, etc.

## Coordinate System

A coordinate system is a reference framework used to represent points or vectors in space. The most commonly used coordinate system is the Cartesian coordinate system.

**Cartesian Coordinate System:**

- The Cartesian coordinate system is based on a set of mutually perpendicular axes: x, y, and z in three dimensions.
- In two dimensions, the axes are x and y, forming a plane.
- Each axis is associated with a number line, where positive values extend in one direction and negative values in the opposite direction.
- A point or vector in the coordinate system is represented by its coordinates (x, y) or (x, y, z) in the respective dimensions.
- The coordinates of a point/vector represent the distances from the origin along each axis.

`Example`:

- Let's consider a two-dimensional vector ùêØ = [3, 4].

- The resulting vector points to the coordinates (3, 4) in the Cartesian plane.
- The magnitude of the vector can be calculated using the Pythagorean theorem: magnitude = ‚àö(3¬≤ + 4¬≤) = 5.

Understanding vector dimensions and the coordinate system is crucial for working with vectors in various mathematical and physical contexts. These concepts allow us to precisely represent and manipulate vectors in different dimensions and understand their geometric properties.

## Basic Vector Operations

There are several fundamental operations that can be performed on vectors, including vector-vector addition, vector-scalar multiplication, and linear combinations.

### 1. Vector-Vector Addition
Vector-vector addition involves adding corresponding elements of two vectors of the same dimension. The result is a new vector with each element being the sum of the corresponding elements from the input vectors.

`Example`:

Let's consider two 3D vectors, $ùêö = [1, 2, 3]$ and $ùêõ = [4, 5, 6]$. Their sum, $ùêö + ùêõ$, can be obtained by adding the corresponding elements:

$ùêö + ùêõ = [1 + 4, 2 + 5, 3 + 6] = [5, 7, 9]$

```python
import numpy as np

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

sum = a + b
print(sum)
```

### 2. Vector-Scalar Multiplication

Vector-scalar multiplication involves multiplying each element of a vector by a scalar value. The result is a new vector with each element scaled by the scalar.

`Example:`

Consider a 2D vector $ùêØ = [3, -2]$. If we multiply it by a scalar $ùëò = 2$, the result would be:

$ùëòùêØ = 2 * [3, -2] = [6, -4]$

import numpy as np

v = np.array([3, -2])
k = 2

result = k * v
print(result)

### 3. Linear Combinations of Vectors

A linear combination of vectors involves scaling each vector by a scalar and then summing them. This operation allows us to create new vectors from existing vectors.

`Example:`

Let's consider two 2D vectors, $ùêö = [1, 2]$ and $ùêõ = [-3, 4]$. We can create a new vector $ùêØ$ as a linear combination of ùêö and ùêõ by multiplying $ùêö$ by $2$ and $ùêõ$ by $-1$, and then adding them:

$ùêØ = 2ùêö - ùêõ = 2 * [1, 2] - [-3, 4] = [5, 0]$

import numpy as np

a = np.array([1, 2])
b = np.array([-3, 4])

v = 2 * a - b
print(v)

### 4. Vector-Vector Multiplication: Dot Product

The dot product of two vectors is a scalar value obtained by multiplying the corresponding elements of the vectors and then summing them. It measures the similarity or correlation between the vectors.

`Example:`

Consider two 3D vectors, $ùêö = [2, 1, -3]$ and $ùêõ = [4, -2, 5]$. The dot product of $ùêö$ and $ùêõ$ can be calculated as:

$ùêö ¬∑ ùêõ = (2 * 4) + (1 * -2) + (-3 * 5) = 8 - 2 - 15 = -9$

import numpy as np

a = np.array([2, 1, -3])
b = np.array([4, -2, 5])

dot_product = np.dot(a, b)
print(dot_product)

## Vector Space, Span, and Subspace

Vector spaces are sets of vectors that satisfy certain properties. The span of a set of vectors is the set of all possible linear combinations of those vectors. A subspace is a subset of a vector space that also satisfies the vector space properties.

### 1. Vector Space

A vector space is a set of vectors closed under vector addition and scalar multiplication. It must satisfy the following properties:

- **Closure under addition:** For any vectors $ùêö$ and $ùêõ$ in the vector space, their sum $ùêö + ùêõ$ is also in the vector space.
- **Closure under scalar multiplication:** For any vector $ùêö$ in the vector space and any scalar $ùëò$, the product $ùëòùêö$ is also in the vector space.
- **Associativity of addition:** For any vectors $ùêö$, $ùêõ$, and $ùëê$ in the vector space, $(ùêö + ùêõ) + ùëê = ùêö + (ùêõ + ùëê)$.
- **Commutativity of addition:** For any vectors $ùêö$ and $ùêõ$ in the vector space, $ùêö + ùêõ = ùêõ + ùêö$.
- **Identity element of addition:** There exists a vector called the zero vector, denoted as $ùüé$, such that for any vector $ùêö$, $ùêö + ùüé = ùêö$.
- **Inverse elements of addition:** For any vector $ùêö$, there exists a vector $ùêõ$ such that $ùêö + ùêõ = ùüé$.
- **Distributivity of scalar multiplication:** For any scalar $ùëò$ and vectors $ùêö$ and $ùêõ$ in the vector space, $ùëò(ùêö + ùêõ) = ùëòùêö + ùëòùêõ$.
- **Distributivity of vector multiplication:** For any scalars $ùëò$ and $ùëô$ and vector $ùêö$ in the vector space, $(ùëò + ùëô)ùêö = ùëòùêö + ùëôùêö$.
- **Compatibility of scalar multiplication with field multiplication:** For any scalars $ùëò$ and $ùëô$ and vector $ùêö$ in the vector space, $(ùëòùëô)ùêö = ùëò(ùëôùêö)$.

### 2. Vector Span

The span of a set of vectors is the set of all possible linear combinations of those vectors. In other words, it is the set of all vectors that can be obtained by scaling each vector in the set and adding them together.

Here are the steps to understand vector span:

- **Definition:** Given a set of vectors $S = {v1, v2, ..., vn}$, the span of $S$, denoted as $Span(S)$, is the set of all possible linear combinations of the vectors in $S$.

- **Linear Combination:** A linear combination of vectors is formed by scaling each vector by a scalar and then adding them together. For example, given vectors $v1 = [1, 2]$ and $v2 = [3, 4]$, a linear combination would be $c1v1 + c2v2$, where $c1$ and $c2$ are scalars.

- `Example:` Let's consider an example with two vectors in $R^2$ (2-dimensional Euclidean space). Suppose we have two vectors $v1 = [1, 2]$ and $v2 = [3, 4]$. We want to find the span of these vectors, i.e., all possible vectors we can obtain by linearly combining them.

- **Span Calculation:** To determine the span, we need to find all possible linear combinations of $v1$ and $v2$. Let's denote a generic vector in the span as $x = [x1, x2]$. We can express $x$ as a linear combination of $v1$ and $v2$ as follows:

$x = c1v1 + c2v2$

Substitute the values of $v1$ and $v2$:

$[x1, x2] = c1[1, 2] + c2[3, 4]$

Distribute the scalars:

$[x1, x2] = [c1, 2c1] + [3c2, 4c2]$

Combine like terms:

$[x1, x2] = [c1 + 3c2, 2c1 + 4c2]$

From this equation, we can see that any vector $[x1, x2]$ in the span can be expressed in terms of $c1$ and $c2$. Therefore, the span of $v1$ and $v2$ is the set of all vectors $[x1, x2]$ satisfying the equation.

Let's implement the example using Python to find the span of $v1$ and $v2$.

```python
import numpy as np
import matplotlib.pyplot as plt

v1 = np.array([1, 2])
v2 = np.array([3, 4])

# Generate all possible linear combinations
span_vectors = []
for c1 in range(-10, 11):
    for c2 in range(-10, 11):
        vector = c1 * v1 + c2 * v2
        span_vectors.append(vector)

# Remove duplicates and convert to NumPy array
span_vectors = np.unique(span_vectors, axis=0)

# Extract x and y coordinates from span vectors
x_coords = span_vectors[:, 0]
y_coords = span_vectors[:, 1]

# Plotting the scatter plot
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.scatter(x_coords, y_coords)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot of Span Vectors')
plt.grid(True)
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')

# Plotting the lines representing the span vectors
plt.subplot(1, 2, 2)
for vector in span_vectors:
    x = [0, vector[0]]
    y = [0, vector[1]]
    plt.plot(x, y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Lines Representing Span Vectors')
plt.grid(True)
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')

plt.tight_layout()
plt.show()
```

The first figure represents the span of the vectors $v1$ and $v2$ as a scatter plot. The span is the set of all possible linear combinations of $v1$ and $v2$. Each point in the scatter plot represents a vector that can be obtained by scaling $v1$ and $v2$ and adding them together. The x-axis represents the first component of the vectors $(x1)$, and the y-axis represents the second component $(x2)$. By plotting these points, we can visualize the entire set of vectors in the span.

The second figure represents the span of the vectors $v1$ and $v2$ as lines connecting the origin $(0, 0)$ to each vector in the span. Each line in the plot represents a vector in the span and indicates the direction and magnitude of that vector. The lines start from the origin $(0, 0)$ and extend to the corresponding vector in the span. The slope of each line represents the scaling factor applied to $v1$ and $v2$ to obtain the corresponding vector in the span. By plotting these lines, we can visually understand the direction and magnitude of each vector in the span.

Both figures help in visualizing the span of the given vectors and provide insights into the possible combinations and variations of vectors that can be obtained by scaling and adding the original vectors.

## Linear Dependence and Independence

Vectors are considered linearly dependent if one of the vectors can be expressed as a linear combination of the others. Otherwise, they are linearly independent.

`Example:`

Consider three 2D vectors, $ùêö = [1, 2], ùêõ = [-3, 6]$, and $ùêú = [4, -8]$. These vectors are linearly dependent because $ùêú$ can be expressed as a linear combination of $ùêö$ and $ùêõ$: $ùêú = 2ùêö - ùêõ$.

## Vector Null Space

The null space of a matrix represents the set of all vectors that, when multiplied by the matrix, result in the zero vector. It is also known as the kernel of the matrix.

`Example:`

Consider a matrix $ùê¥ = [[1, 2, 3], [2, 4, 6], [3, 6, 9]]$. The null space of $ùê¥$ contains all vectors of the form $[x, y, z]$ that satisfy the equation $ùê¥[x, y, z] = [0, 0, 0]$. In this case, the null space is the span of the vector $[1, -2, 1]$.

## Vector Norms

Norms provide a measure of the length or magnitude of a vector. There are different types of norms, including the Euclidean norm ($L2$), Manhattan norm ($L1$), and Max norm ($L‚àû$).

### 1. Euclidean Norm: L2 Norm

The $L2$ norm, also known as the Euclidean norm or the 2-norm, calculates the square root of the sum of the squares of the vector's elements. It represents the Euclidean distance from the origin to the vector in a geometric sense.

- `Example:`

Consider a 3D vector $ùêØ = [3, -2, 5]$. The $L2$ norm of $ùêØ$ can be calculated as:

$‚ÄñùêØ‚Äñ‚ÇÇ = ‚àö(3¬≤ + (-2)¬≤ + 5¬≤) = ‚àö(9 + 4 + 25) = ‚àö38 ‚âà 6.164$

```python
import numpy as np

v = np.array([3, -2, 5])

l2_norm = np.linalg.norm(v)
print(l2_norm)
```

### 2. Manhattan Norm: L1 Norm

The L1 norm, also known as the Manhattan norm or the 1-norm, calculates the sum of the absolute values of the vector's elements. It represents the distance between two points measured along the axes, similar to how one would navigate through a city by moving along the streets.

`Example:`

Consider a 3D vector $ùêØ = [3, -2, 5]$. The L1 norm of ùêØ can be calculated as:

$‚ÄñùêØ‚Äñ‚ÇÅ = |3| + |-2| + |5| = 3 + 2 + 5 = 10$

```python
import numpy as np

v = np.array([3, -2, 5])

l1_norm = np.linalg.norm(v, ord=1)
print(l1_norm)
```

### 3. Max Norm: L‚àû Norm:

The L‚àû norm, also known as the max norm or the infinity norm, calculates the maximum absolute value of the vector's elements. It represents the maximum distance between two points along any coordinate axis.

`Example:`

Consider a 3D vector $ùêØ = [3, -2, 5]$. The L‚àû norm of $ùêØ$ can be calculated as:

$‚ÄñùêØ‚Äñ‚àû = max(|3|, |-2|, |5|) = 5$

```python
import numpy as np

v = np.array([3, -2, 5])

l_inf_norm = np.linalg.norm(v, ord=np.inf)
print(l_inf_norm)
```

## Vector Inner Product (Dot Product):
The inner product, also known as the dot product, is an operation defined between two vectors that yields a scalar value. It measures the similarity or alignment between the vectors.

- **Definition:** The dot product of two vectors $u = [u1, u2, ..., un]$ and $v = [v1, v2, ..., vn]$ is given by the formula: $u ¬∑ v = u1v1 + u2v2 + ... + unvn$.

- **Geometric Interpretation:** The dot product is related to the angle between vectors. If the dot product of two vectors is positive, the vectors are pointing in similar directions. If it is negative, the vectors are pointing in opposite directions. If it is zero, the vectors are orthogonal (perpendicular) to each other.

`Example:`

Let's consider two vectors $u = [3, 4]$ and $v = [-2, 6]$.

To calculate the dot product:

$u ¬∑ v = (3)(-2) + (4)(6) = -6 + 24 = 18$

Therefore, the dot product of $u$ and $v$ is $18$.

## Vector Length (Norm)

The length, also known as the norm or magnitude, of a vector measures its "size" or magnitude. It quantifies the distance from the origin to the vector's tip. There are different types of vector norms, but the most commonly used is the Euclidean norm.

- **Definition:** The Euclidean norm (length) of a vector $v = [v1, v2, ..., vn]$ is given by the formula: 

$||v|| = \sqrt{v1^2 + v2^2 + ... + vn^2}$

- **Geometric Interpretation:** The length of a vector represents the distance between the origin and the point corresponding to the vector's tip in n-dimensional space.

`Example:`

Consider a vector $v = [3, 4]$.

To calculate the length (norm):

$||v|| = \sqrt{3^2 + 4^2} = \sqrt{9 + 16} = \sqrt{25} = 5$

Therefore, the length of vector $v$ is $5$.

## Vector Distance

The distance between two vectors quantifies the separation or dissimilarity between them. It can be computed using the vector difference and the vector length.

- **Definition:** The distance between two vectors $u$ and $v$ is given by the formula: 

$d(u, v) = ||u - v||$

where $||u - v||$ represents the length of the vector obtained by subtracting $v$ from $u$.

- **Geometric Interpretation:** The distance between two vectors represents the length of the shortest path connecting their tips in n-dimensional space.

`Example:`

Consider two vectors $u = [1, 2, 3]$ and $v = [4, 5, 6]$.

To calculate the distance:

$d(u, v) = ||u - v|| = ||[1-4, 2-5, 3-6]|| = ||[-3, -3, -3]||$

Now, calculate the length of the vector $[-3, -3, -3]$:

$||[-3, -3, -3]|| = \sqrt{(-3)^2 + (-3)^2 + (-3)^2} = \sqrt{9 + 9 + 9} = \sqrt{27}$

Therefore, the distance between $u$ and $v$ is $\sqrt{27}$.

## Vector Angles and Orthogonality

### 1. Vector Angles

Vector angles refer to the angles formed between two vectors in n-dimensional space. Understanding vector angles is essential in various applications, such as geometry, physics, computer graphics, and machine learning.

- **Calculating the Angle Between Two Vectors:** The angle $Œ∏$ between two vectors $u$ and $v$ can be calculated using the dot product and vector norms as follows:

\begin{equation}
Œ∏ = arccos((u ¬∑ v) / (||u|| ||v||))
\end{equation}

- **Geometric Interpretation:** The angle between two vectors determines their degree of alignment or deviation. If the angle is $0¬∞$, the vectors are parallel. If the angle is $90¬∞$, the vectors are orthogonal (perpendicular). If the angle is greater than $90¬∞$, the vectors are in opposite directions.

`Example:`

Let's consider two vectors $u = [3, 4]$ and $v = [-2, 6]$.

To calculate the angle between $u$ and $v$:

$Œ∏ = \arccos{\frac{u ¬∑ v}{||u|| * ||v||}}$

$Œ∏ = arccos{\frac{3*-2 + 4*6}{\sqrt{3^2 + 4^2} * \sqrt{(-2)^2 + 6^2}}}$

```python
import numpy as np

u = np.array([3, 4])
v = np.array([-2, 6])

angle_rad = np.arccos(np.dot(u, v) / (np.linalg.norm(u) * np.linalg.norm(v)))
angle_deg = np.degrees(angle_rad)

print("Angle between u and v:", angle_deg, "degrees")
```

### 2. Orthogonality

Orthogonality refers to the property of vectors being perpendicular or orthogonal to each other. Orthogonal vectors have a dot product of $0$, indicating no alignment or similarity between them.

- **Checking for Orthogonality:** Two vectors $u$ and $v$ are orthogonal if their dot product is $0$:

\begin{equation}
u ¬∑ v = 0
\end{equation}

- **Geometric Interpretation:** Orthogonal vectors are perpendicular to each other and form a right angle. They have no alignment or similarity in their direction.

- `Example:` Let's consider two vectors $u = [3, 4]$ and $v = [-2, 6]$.

- **To check for orthogonality:** Calculate the dot product of $u$ and $v$:

\begin{equation}
u ¬∑ v = (3*-2 + 4*6)
\end{equation}

```python
u = np.array([3, 4])
v = np.array([-2, 6])

dot_product = np.dot(u, v)

if dot_product == 0:
    print("u and v are orthogonal")
else:
    print("u and v are not orthogonal")
```

## Systems of Linear Equations

Vectors can be used to represent systems of linear equations. A system of equations can be written in matrix form as $ùê¥ùê± = ùêõ$, where $ùêÄ$ is the coefficient matrix, $ùê±$ is the vector of variables, and $ùêõ$ is the vector of constants. Solving the system involves finding the values of $ùê±$ that satisfy the equation.

`Example:` Consider the system of equations:
$2ùë• + 3ùë¶ = 8$

$‚àíùë• + 4ùë¶ = 7$

We can write this system in matrix form as:

\begin{align*}
\begin{bmatrix}
2 & 3 \\
-1 & 4 \\
\end{bmatrix}
\begin{bmatrix}
x \\
y \\
\end{bmatrix}
=
\begin{bmatrix}
8 \\
7 \\
\end{bmatrix}
\end{align*}

```python
import numpy as np

A = np.array([[2, 3], [-1, 4]])
b = np.array([8, 7])

x = np.linalg.solve(A, b)
print(x)
```

The solution $ùê±$ represents the values of $ùë•$ and $ùë¶$ that satisfy the system of equations. In this case, $ùë• ‚âà 1.462$ and $ùë¶ ‚âà 1.538$.
